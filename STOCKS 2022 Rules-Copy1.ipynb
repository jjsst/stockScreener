```python

#importing packages
import pandas as pd
import numpy as np
import datetime as dt
from datetime import datetime
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import bs4 as bs
import requests
import os
from yahoo_fin import stock_info as si
import talib as tlib
import pandas_ta as taa
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
from docx import Document
from docx.shared import Cm , Pt
from pandas import ExcelWriter
from openpyxl import load_workbook
from ta.utils import dropna
from ta.volatility import BollingerBands
import re
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
import seaborn as sns

sns.set()
pd.set_option('display.max_colwidth', 25)
pd.set_option('display.precision', 2)
pd.set_option('display.float_format', '{:.2f}'.format)

tickers = []

excel_stock_ark_data_df = pd.read_excel('Stocks.xlsx', sheet_name='sheet_stocks')

tickers = excel_stock_ark_data_df['ticker'].tolist()

all_data = pd.DataFrame()
test_data = pd.DataFrame()
test_fin = pd.DataFrame()

no_data = []

#Extract data from Yahoo Finance


cnt=0
for i in tickers:

    cnt = cnt + 1
    print(str(i) + ' nº ' + str(cnt))
    test_data = yf.download(i, start=dt.datetime(2020,1,1), end = dt.date.today(), progress=False)  
    test_data['symbol'] = i

#     try:        
#         objstock = yf.Ticker(i)
#         info = objstock.info
#         test_data['sector'] = info['sector']     
#         test_data['name'] = info['shortName']              
#     except Exception as e:
#         print(e)  

    open = test_data['Open']
    high  = test_data['High']
    low  = test_data['Low']
    close  = test_data['Close']
    volume = test_data['Volume']
    test_data['ema20'] = tlib.EMA(close, timeperiod=20)
    test_data['ema20'] = test_data['ema20'].round(decimals = 2)
    test_data['rsi14'] = tlib.RSI(close, timeperiod=14)
    test_data['rsi14'] = test_data['rsi14'].round(decimals = 2)
    test_data['mfi12'] = tlib.MFI(high, low, close, volume, timeperiod=12)
    test_data['mfi12'] = test_data['mfi12'].round(decimals = 2)    
    test_data['cdle_hammer'] = tlib.CDLHAMMER(open, high, low, close)   
    test_data['cdle_haramicross'] = tlib.CDLHARAMICROSS(open, high, low, close)
    test_data['cdle_piercing'] = tlib.CDLPIERCING(open, high, low, close)
    test_data['cdle_engulfing'] = tlib.CDLENGULFING(open, high, low, close)     
    test_data['cdle_3blackcrows'] = tlib.CDL3BLACKCROWS(open, high, low, close)
    test_data['cdle_darkcloudcover'] = tlib.CDLDARKCLOUDCOVER(open, high, low, close, penetration=0)
    test_data['cdle_hangingman'] = tlib.CDLHANGINGMAN(open, high, low, close)
    test_data['cdle_eveningstar'] = tlib.CDLEVENINGSTAR(open, high, low, close, penetration=0)
    test_data["ValHigh_of_365Days"]= tlib.MAX(close, timeperiod=365)
    test_data["DayHigh_of_365Days"]= tlib.MAXINDEX(close, timeperiod=365)
    test_data["ValLow_of_365Days"]= tlib.MIN(close, timeperiod=365)
    test_data["DayLow_of_365Days"]= tlib.MININDEX(close, timeperiod=365)    

    
    test_data = test_data.dropna().copy()
    all_data = all_data.append(test_data)


all_data_current = all_data.groupby('symbol').tail(1)

# escreve para ficheiro
data = datetime.today().strftime('%Y-%m-%d')
all_data_current.to_excel(r'AnaliseDiaria '+data+ '.xlsx', index = True)  # para copiar a data coluna index

```
